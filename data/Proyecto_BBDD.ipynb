```python
import pandas as pd
import psycopg2
from psycopg2.extras import execute_values
```

```python
df1 = pd.read_csv("clase_1.csv",sep=';')
df2 = pd.read_csv("clase_2.csv",sep=';')
df3 = pd.read_csv("clase_3.csv",sep=';')
df4 = pd.read_csv("clase_4.csv",sep=';')

```

```python
df1.head()
```

```python
df1.columns = df1.columns.str.strip()  # Elimina espacios en los nombres
df1.columns = df1.columns.str.replace(" ", "_")  # Sustituye espacios por guiones bajos
df2.columns = df2.columns.str.strip()  # Elimina espacios en los nombres
df2.columns = df2.columns.str.replace(" ", "_")  # Sustituye espacios por guiones bajos
df3.columns = df3.columns.str.strip()  # Elimina espacios en los nombres
df3.columns = df3.columns.str.replace(" ", "_")  # Sustituye espacios por guiones bajos
df4.columns = df4.columns.str.strip()  # Elimina espacios en los nombres
df4.columns = df4.columns.str.replace(" ", "_")  # Sustituye espacios por guiones bajos
```

```python
dfalumnos1 = df1[["Nombre", "Email", "Fecha_comienzo"]]
dfalumnos2 = df2[["Nombre", "Email", "Fecha_comienzo"]]
dfalumnos3 = df3[["Nombre", "Email", "Fecha_comienzo"]]
dfalumnos4 = df4[["Nombre", "Email", "Fecha_comienzo"]]

dfpromocion1 = df1[["Promoción"]]
dfpromocion2 = df2[["Promoción"]]
dfpromocion3 = df3[["Promoción"]]
dfpromocion4 = df4[["Promoción"]]

dfalumnos1["Fecha_comienzo"] = pd.to_datetime(dfalumnos1["Fecha_comienzo"], format="%d/%m/%Y").dt.strftime("%Y-%m-%d")
dfalumnos2["Fecha_comienzo"] = pd.to_datetime(dfalumnos2["Fecha_comienzo"], format="%d/%m/%Y").dt.strftime("%Y-%m-%d")
dfalumnos3["Fecha_comienzo"] = pd.to_datetime(dfalumnos3["Fecha_comienzo"], format="%d/%m/%Y").dt.strftime("%Y-%m-%d")
dfalumnos4["Fecha_comienzo"] = pd.to_datetime(dfalumnos4["Fecha_comienzo"], format="%d/%m/%Y").dt.strftime("%Y-%m-%d")

dfalumnos = pd.concat([dfalumnos1,dfalumnos2,dfalumnos3,dfalumnos4], ignore_index= True)
dfpromocion = pd.concat([dfpromocion1,dfpromocion2,dfpromocion3,dfpromocion4], ignore_index= True)

```
